## Concepts and terminologies

1. Hypothesis



```python
#@title
# ✓It is an assertion (claim, assumption, hypothesis) made 
# about a value of a population parameter
# Population parameters: Mean, Median, Proportion,Standard deviation, etc.
```

2. Hypothesis testing

```python
#@title
# testing a statistical hypothesis (or claim) to determine 
# if the claim is true or not and determine to which extent 
# (Probability, C.I,significance level) it is true (No absolute certainty).
```

3. Null and alternate hypothesis

```python
#@title
# ✓The null hypothesis (H0) is a statement concerning the value of the
# parameter of interest (θ) in the claim that is made. It is in the form
# ✓ H0: θ = θ0 or H0: θ1 - θ2 = θ0

# ✓The alternative hypothesis (H1) is a statement about the parameter
# (θ) that is believed to be true if (H0) is not true. (H1) can be one of
# these:
# ✓H1: θ ≠ θ0 (2 Tails) or H1: θ < θ0 (1 Tail) or H1: θ > θ0 (1T)
# ✓Or
# ✓H1: θ1 - θ2 ≠ θ0 (2T) or H1: θ1 - θ2 < θ0 (1T) or H1: θ1 - θ2 > θ0 (1T)
```

5. P-value

```python
#@title
# is the lowest (smallest) level of significance at which Ho
# can be rejected.
# If p<α, we reject the null hypothesis.
# If p≥α, we fail to reject the null hypothesis.
```

6. Significance level, $\alpha$

```python
#@title
# we have to determine when do we say an observation is 
# statistically significant.

# α is a measure of the strength of the evidence that must 
# be present in your sample before you will reject 
# the null hypothesis and conclude that the effect is 
# statistically significant. 
```
