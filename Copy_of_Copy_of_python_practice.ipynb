```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(4,2))
sns.countplot(x = data.smoker);
```

```python
plt.figure(figsize=(4,2))
sns.countplot(y = data.day, color="blue");
```

```python
count = order=data['day'].value_counts().index
plt.figure(figsize=(4,2))
sns.countplot(y = data.day, order =count, color="r");
```

#  sorting in Descending

```python
plt.figure(figsize=(4,2))
sns.countplot(y = data.day, order=data['day'].value_counts().sort_values(ascending=True).index, color="r");
```

```python
#visualizing numerious features|total bills and tips

fig, axes = plt.subplots(nrows = 1,ncols = 2, figsize=(15,3))
sns.histplot(data["total_bill"], ax=axes[0]);
sns.histplot(data["tip"], ax=axes[1]);
```

Bivariate Analysis

```python
#total_bills by sex
total_billsex = data.groupby("sex")["total_bill"].sum().reset_index()
total_billsex
```

```python
total_billtip = data.groupby("sex")["tip"].sum().reset_index()
total_billtip
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 2, figsize=(15,3))
sns.barplot(x = "sex", y = "total_bill", data = total_billsex, ax=axes[0])
ax=axes[0].set_title("yyy")
sns.barplot(x = "sex", y = "tip", data = total_billtip, ax=axes[1])
ax=axes[1].set_title("xx")
```
