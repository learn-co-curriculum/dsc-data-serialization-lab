###   What loss function are we optimising?

> Binary or Multi-class Cross entropy:

$Binary = -(p(x)\cdot \log q(x) + (1-p(x)) \cdot \log (1-q(x))$

$ Multi-class = -\sum_x p(x) \cdot q(x)$

Reference: [Cross Entropy](https://towardsdatascience.com/cross-entropy-for-classification-d98e7f974451)
